# Stacked bar chart preprocessing

```python
import pandas as pd
import json
```

```python
df = pd.read_excel('raw_data/Europe-Central-Asia_aggregated_data_up_to-2026-01-10.xlsx')
```

```python
df = df[df['COUNTRY']=='Ukraine']
df = df[df['EVENT_TYPE']!='Strategic developments']
```

```python
df['WEEK'] = pd.to_datetime(df['WEEK'])
df['YEAR'] = df['WEEK'].dt.year
df = df[df['YEAR'] < 2026]
```

```python
df.head()
```

```python
df_clean = df.groupby(['YEAR', 'EVENT_TYPE'])['EVENTS'].sum().reset_index()
df_clean = df_clean.rename(columns={'EVENTS': 'count'})
df_clean.to_json('../../data/final/stackedbarchart/Year_Events_UKR_v2.json', orient='records')
```
