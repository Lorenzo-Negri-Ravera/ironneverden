```python
import pandas as pd
```

```python
path_acled  = "raw_data/Europe-Central-Asia_aggregated_data_up_to-2026-01-10.xlsx"
data_acled = pd.read_excel(path_acled)
data_acled.head()
```

```python
data_acled = data_acled[data_acled["COUNTRY"] == "Ukraine"]
data_acled['YEAR'] = pd.to_datetime(data_acled['WEEK']).dt.year
```

```python
data_sub_event = data_acled.groupby(["WEEK", 'YEAR', 'EVENT_TYPE', 'SUB_EVENT_TYPE'])['EVENTS'].sum().reset_index()

```

```python

data_sub_evens_ukr = data_sub_event.drop(["EVENT_TYPE"], axis = 1).reset_index(drop = True)
sub_event_type_events = data_sub_evens_ukr[(data_sub_evens_ukr["SUB_EVENT_TYPE"] == "Shelling/artillery/missile attack") |
                                           (data_sub_evens_ukr["SUB_EVENT_TYPE"] == "Air/drone strike") |
                                           (data_sub_evens_ukr["SUB_EVENT_TYPE"] == "Armed clash")]
sub_event_linechart = sub_event_type_events[sub_event_type_events["YEAR"] >= 2022].reset_index(drop = True)
sub_event_linechart['WEEK'] = sub_event_linechart['WEEK'].dt.strftime('%Y-%m-%d')
```

```python
sub_event_linechart.to_json("../../data/final/events_line_chart/events_line_chart_v2.json", orient='records')
```
