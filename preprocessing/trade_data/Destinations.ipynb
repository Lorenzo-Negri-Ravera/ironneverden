```python
# Libraries
import pandas as pd
```

```python
# Datasets
RUS_export_2021 = pd.read_csv('raw_data/OEC_data/2021_Russia_exports_destinations.csv')
RUS_export_2023 = pd.read_csv('raw_data/OEC_data/2023_Russia_exports_destinations.csv')
UKR_export_2021 = pd.read_csv('raw_data/OEC_data/2021_Ukraine_exports_destinations.csv')
UKR_export_2023 = pd.read_csv('raw_data/OEC_data/2023_Ukraine_exports_destinations.csv')
```

```python
# Cleaning: removal of 'Country ID', 'Continent ID', 'Year'
RUS_export_2021 = RUS_export_2021.drop(['Country ID', 'Continent ID', 'Year'], axis=1)
RUS_export_2023 = RUS_export_2023.drop(['Country ID', 'Continent ID', 'Year'], axis=1)
UKR_export_2021 = UKR_export_2021.drop(['Country ID', 'Continent ID', 'Year'], axis=1)
UKR_export_2023 = UKR_export_2023.drop(['Country ID', 'Continent ID', 'Year'], axis=1)
```

```python
# Set values of column 'ISO 3' in Uppercase mode
RUS_export_2021['ISO 3'] = RUS_export_2021['ISO 3'].str.upper()
RUS_export_2023['ISO 3'] = RUS_export_2023['ISO 3'].str.upper()
UKR_export_2021['ISO 3'] = UKR_export_2021['ISO 3'].str.upper()
UKR_export_2023['ISO 3'] = UKR_export_2023['ISO 3'].str.upper()
```

```python
# Export the dataframes into jason files
RUS_export_2021.to_json('../../data/final/trade-data/final_datasets/RUS_export_2021_destinations_v2.json', orient='records')
RUS_export_2023.to_json('../../data/final/trade-data/final_datasets/RUS_export_2023_destinations_v2.json', orient='records')
UKR_export_2021.to_json('../../data/final/trade-data/final_datasets/UKR_export_2021_destinations_v2.json', orient='records')
UKR_export_2023.to_json('../../data/final/trade-data/final_datasets/UKR_export_2023_destinations_v2.json', orient='records')
```
