```python
import pandas as pd
```

```python
path_ukr  = "../../preprocessing/spike_map/raw_data/ACLED Data_2025-12-30_UKR.csv"
```

```python
data_ukr = pd.read_csv(path_ukr, sep=';', engine='python')  
```

```python
df_ukr = data_ukr[data_ukr["iso"] == 804]
```

```python
columns=  ['event_id_cnty','event_date','disorder_type',
            'event_type', 'sub_event_type','interaction', 
            'iso', 'admin1', 'admin2', 'admin3', 'location', 
            'latitude', 'longitude','fatalities',"year"]
```

```python
df_sp_map_ukr = df_ukr[columns]
```

```python
spikemap_ukr_regions    = df_sp_map_ukr.groupby(["admin1", "event_type","year"])["fatalities"].sum().reset_index()
spikemap_ukr_municipals = df_sp_map_ukr.groupby(["admin3", "event_type","year"])["fatalities"].sum().reset_index()
spikemap_ukr_districts  = df_sp_map_ukr.groupby(["admin2", "event_type","year"])["fatalities"].sum().reset_index()
```

```python
spikemap_ukr_regions    .to_json("../data/final/spike/regions/spikemap_ukr_regions.json"       , orient="records",  force_ascii=False, indent=4)             
spikemap_ukr_municipals .to_json("../data/final/spike/municipals/spikemap_ukr_municipals.json" , orient="records",  force_ascii=False, indent=4)  
spikemap_ukr_districts  .to_json("../data/final/spike/regions/spikemap_ukr_dinsticts.json"     , orient="records",  force_ascii=False, indent=4)  
```
