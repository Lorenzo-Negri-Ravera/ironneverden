```python
import pandas as pd
```

```python
path_ukr  = "raw_data/ACLED Data_2025-12-30_UKR.csv"
```

```python
data_ukr = pd.read_csv(path_ukr, sep=';', engine='python')  
```

```python
df_ukr = data_ukr[data_ukr["iso"] == 804]
df_ukr.head()
```

```python
columns=  ['event_id_cnty','event_date', 'year','disorder_type',
            'event_type', 'sub_event_type','interaction', 
            'iso', 'admin1', 'admin2', 'admin3', 'location', 
            'latitude', 'longitude','fatalities']
```

```python
df_sp_map_ukr = df_ukr[columns]
```

```python
df_sp_map_ukr = df_sp_map_ukr[df_sp_map_ukr['event_type'].isin(['Battles', 'Explosions/Remote violence'])].reset_index(drop=True)
df_sp_map_ukr.head()
```

```python
df_sp_map_ukr_pre2022 = df_sp_map_ukr[df_sp_map_ukr["year"] < 2022]
df_sp_map_ukr_pre2022.head()
```

```python
spikemap_ukr_regions    = df_sp_map_ukr.groupby(["admin1", "event_type"])["fatalities"].sum().reset_index()
spikemap_ukr_municipals = df_sp_map_ukr.groupby(["admin3", "event_type", "year"])["fatalities"].sum().reset_index()
spikemap_ukr_districts  = df_sp_map_ukr.groupby(["admin2", "event_type"])["fatalities"].sum().reset_index()
```

```python
spikemap_ukr_regions    .to_csv("../../data/final/spike/region/spikemap_ukr_regions_v2.csv"       , index=False)
spikemap_ukr_municipals .to_csv("../../data/final/spike/municipals/spikemap_ukr_municipals_v2.csv" , index=False)
spikemap_ukr_districts  .to_csv("../../data/final/spike/district/spikemap_ukr_districts_v2.csv"   , index=False)
```
