```python

def find_nation(incao_code):
    # Usa .get() invece di []. 
    # Se 'incao_code' non esiste nel dizionario, restituisce None invece di dare errore.
    airport_data = airports.get(incao_code)
    
    if airport_data:
        return airport_data.get('country')
    else:
        return None # Oppure puoi mettere stringa "Unknown"

def load_data(path, colonne):
    DATA = []
    for data in path:
        df_temp = pd.read_csv(data, usecols = colonne)
        DATA.append(df_temp)
        
    print(len(DATA))
    df_concat = pd.concat(DATA, ignore_index=True)

    df_concat['destination_name'] = df_concat['destination'].map(find_nation,na_action='ignore')
    df_concat['origin_name'] = df_concat['origin'].map(find_nation, na_action='ignore')
    df_concat = df_concat.drop("destination", axis = 1)
    df_concat = df_concat.drop("origin", axis = 1)
    
    print(f"Total_row: {len(df_concat)}, added destination_name")
    
    return df_concat

def load_data2(path):
    DATA = []
    for data in path:
        df_temp = pd.read_csv(data, usecols = colonne)
        DATA.append(df_temp)
        
    print(len(DATA))
    df_concat = pd.concat(DATA, ignore_index=True)
    print(f"Total_row: {len(df_concat)}, added destination_name")
    
    return df_concat
```

```python
# import pandas as pd

def monthly_origin_counts(df, day_col='day', orig_col='origin_name', 
                          utc=True, period='M', as_pivot=True, 
                          save_path=None, verbose=False):
    """
    Raggruppa i voli per mese (YYYY-MM) e ORIGINE.
    Restituisce (monthly_orig, monthly_pivot) â€” monthly_pivot puÃ² essere None se as_pivot=False.
    """
    df2 = df.copy()
    if not pd.api.types.is_datetime64_any_dtype(df2[day_col]):
        df2[day_col] = pd.to_datetime(df2[day_col], utc=utc, errors='coerce')
    df2 = df2.dropna(subset=[day_col, orig_col])
    df2['month'] = df2[day_col].dt.to_period(period).astype(str)
    monthly_orig = (df2.groupby(['month', orig_col])
                         .size()
                         .reset_index(name='flights'))
    monthly_pivot = None
    if as_pivot:
        monthly_pivot = (monthly_orig
                         .pivot(index=orig_col, columns='month', values='flights')
                         .fillna(0)
                         .astype(int))
    if save_path:
        monthly_orig.to_csv(save_path, index=False)
    return monthly_orig, monthly_pivot




def monthly_destination_counts(df, day_col='day', dest_col='destination_name',
                               utc=True, period='M', as_pivot=True,
                               save_path=None, verbose=False):
    """
    Raggruppa i voli per mese (YYYY-MM) e destinazione.
    Restituisce (monthly_dest, monthly_pivot) â€” monthly_pivot puÃ² essere None se as_pivot=False.
    """
    df2 = df.copy()
    # assicurati datetime
    if not pd.api.types.is_datetime64_any_dtype(df2[day_col]):
        df2[day_col] = pd.to_datetime(df2[day_col], utc=utc, errors='coerce')
    # rimuovi righe senza giorno o destinazione validi
    df2 = df2.dropna(subset=[day_col, dest_col])
    df2['month'] = df2[day_col].dt.to_period(period).astype(str)

    monthly_dest = (df2.groupby(['month', dest_col])
                        .size()
                        .reset_index(name='flights'))

    monthly_pivot = None
    if as_pivot:
        monthly_pivot = (monthly_dest
                         .pivot(index=dest_col, columns='month', values='flights')
                         .fillna(0)
                         .astype(int))

    if save_path:
        monthly_dest.to_csv(save_path, index=False)

    # if verbose:
    #     print(monthly_dest.head())

    return monthly_dest, monthly_pivot




def monthly_origin_dest_counts(df, day_col='day', orig_col='origin_name', dest_col='destination_name', utc=True, period='M', save_path=None, verbose=False):
    """
    Raggruppa i voli per mese (YYYY-MM), ORIGINE e DESTINAZIONE.
    Restituisce il dataframe piatto (non pivot, perchÃ© con 3 variabili il pivot Ã¨ complesso).
    """
    df2 = df.copy()
    if not pd.api.types.is_datetime64_any_dtype(df2[day_col]):
        df2[day_col] = pd.to_datetime(df2[day_col], utc=utc, errors='coerce')
    df2 = df2.dropna(subset=[day_col, orig_col, dest_col])
    df2['month'] = df2[day_col].dt.to_period(period).astype(str);
    monthly_counts = (df2.groupby(['month', orig_col, dest_col]).size().reset_index(name='flights'))
    if save_path:
        monthly_counts.to_csv(save_path, index=False)
    return monthly_counts


```

```python
# i = 1

def RigheUtili(data):
    data_V = data[(data["flights_2022"] != 0) & (data["flights_2022"] != 0)]
    data_V = data_V[(data_V["flights_2022"] >= 100) | (data_V["flights_2022"] >= 100)]
    data_V["Divergence"] = data_V["flights_2022"] - data_V["flights_2021"]
    data_V["Diff 2022-2021%"] = ((data_V["flights_2022"] - data_V["flights_2021"])/(data_V["flights_2021"])).round(2)
    data_V["Absol. diff. 2021-2022"] = (data_V["flights_2022"] - data_V["flights_2021"])
    data_V = data_V.rename(columns = { "mese_id": "month","flights_2021": "2021","flights_2022": "2022"})
    
    return data_V
```
